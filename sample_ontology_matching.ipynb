# --- 2. Load Sentence Embedding Model ---

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

def embed_ontology(ontology):
    """
    Given an ontology (dictionary of concept: definition),
    return a dictionary of concept: embedding.
    """
    embeddings = {}
    for concept, definition in ontology.items():
        text = f"{concept} - {definition}"
        # Generate an embedding for the combined text
        embedding = model.encode(text, convert_to_tensor=True)
        embeddings[concept] = embedding
    return embeddings

llm_embeddings = embed_ontology(sample_llm_ontology)
domain_embeddings = embed_ontology(sample_domain_ontology)
```

# --- 3. Match Concepts Based on Cosine Similarity ---


```python
similarity_threshold = 0.55
accepted_matches = []
unmatched_concepts = []

for llm_concept, llm_emb in llm_embeddings.items():
    best_match = None
    best_score = 0
    # Compare against every domain concept
    for domain_concept, domain_emb in domain_embeddings.items():
        score = util.cos_sim(llm_emb, domain_emb).item()
        if score > best_score:
            best_score = score
            best_match = domain_concept

    if best_score >= similarity_threshold:
        accepted_matches.append((llm_concept, best_match, best_score))
        print(f"✅ Matched '{llm_concept}' with '{best_match}' (Similarity: {best_score:.2f})")
    else:
        unmatched_concepts.append(llm_concept)
        print(f"❌ No strong match for '{llm_concept}' (Best Similarity: {best_score:.2f})")

print("\nSummary:")
print(f"Total LLM Concepts: {len(llm_embeddings)}")
print(f"Matched Concepts: {len(accepted_matches)}")
print(f"Unmatched Concepts: {len(unmatched_concepts)}")
```
